```python
loan_df['Loan_Status']=loan_df['Loan_Status'].map({'N':0,'Y':1})
```

```python
imputer = KNNImputer(n_neighbors= 3)
impute_with_3 = imputer.fit_transform(loan_df[['Loan_Status','Gender']])

print("\nImpute with 2 Neighbour: \n", impute_with_3)
```
