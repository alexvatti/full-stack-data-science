# Connecting - local MS -SQL SERVER using Python
# Run the SQL Query on the Table

```python
import pyodbc

# Establish connection
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=ANTONY\SQLEXPRESS;DATABASE=MOVIESDB;Trusted_Connection=yes;')

# Create a cursor
cursor = conn.cursor()

# Execute SQL query
query_list = ["SELECT * FROM BirdDetails;", "SELECT bird_name , age  FROM BirdDetails;","SELECT *  FROM BirdDetails WHERE visits < 4;"]


for query in query_list:
    print("cmd : ", query)
    print("========================================")
    cursor.execute(query)
    # Fetch data
    for row in cursor:
        print(row)
    print("==========================================")

# Close cursor and connection
cursor.close()
conn.close()
```

# Connect the LOCAL MS-SQL SERVER
# Read the table - then create data frame

```python
import pyodbc
import pandas as pd
import warnings
warnings.filterwarnings("ignore")


# Establish a connection to the database
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=ANTONY\SQLEXPRESS;DATABASE=MOVIESDB;Trusted_Connection=yes;')

# Define the SQL query to retrieve data from the table
sql_query = "SELECT * FROM BirdDetails"

# Execute the query and fetch the results into a Pandas DataFrame
df = pd.read_sql(sql_query, conn)

# Close the database connection
conn.close()

# Display the DataFrame
print(df)
```
