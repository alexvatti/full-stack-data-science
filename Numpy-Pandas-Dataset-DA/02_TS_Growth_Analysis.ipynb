## Import Python Modules

```python
# Load necessary pyhton modules
import numpy as np
import pandas as pd
```

## Load the Dataset

```python
df_districts = pd.read_csv("dataset/C7/dataset/dim_districts.csv")
df_date = pd.read_csv("dataset/C7/dataset/dim_date.csv")
df_transport = pd.read_csv("dataset/C7/dataset/fact_transport.csv")
```

## Basic Inspection on dataset

```python
def basic_inspection(table):
    print("Top 5 Records of dataset")
    print(table.head())
    print()
        
    print("Bottom Records of dataset")
    print(table.tail())
    print()
    
    print("Column/features/Variable  - Names of Given dataset")
    print(table.columns)
    print()
    
    print("Shape(rows x columns) - of Given dataset")
    print(table.shape)
    print()
    
    print("Data types - Given Column Names")
    print(table.dtypes)
    print()
    
    print("Summry of dataset")
    print(table.info())
    print()
    
    print("To see the count of null/nan values in columns of dataset")
    print(table.isnull().value_counts())
    print()
    
    print("Dataset Summary ")
    print(table.describe())
    print()
    
```

```python
basic_inspection(df_transport)
```

```python
districts_dict = df_districts.set_index('dist_code')['district'].to_dict()
```

```python
result_date_dict = df_date.set_index('date')[['Month', 'quarter', 'fiscal_year']].to_dict(orient='index')
```

```python
result_date_dict
```

```python
df_transport['month'] = pd.to_datetime(df_transport['month'])
```

```python
df_transport['Date'] = df_transport['month'].dt.strftime('%d-%m-%Y')
```

```python
df_transport['Month'] = df_transport['Date'].map(lambda x: result_date_dict[x]['Month'])
df_transport['Quarter'] = df_transport['Date'].map(lambda x: result_date_dict[x]['quarter'])
df_transport['Fiscal_Year'] = df_transport['Date'].map(lambda x: result_date_dict[x]['fiscal_year'])
```

```python
df_transport.head()
```

```python
df_transport=df_transport.drop(columns=["month","Date"])
```
