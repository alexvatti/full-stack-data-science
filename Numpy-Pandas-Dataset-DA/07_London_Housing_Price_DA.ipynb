## Import Python Modules

```python
# Load necessary python modules
import numpy as np
import pandas as pd
```

## Load Dataset

```python
df = pd.read_csv("dataset/London_Housing_Price_dataset.csv")
```

## Basic Inspection on dataset

```python
def basic_inspection(table):
    print("Top 5 Records of dataset")
    print(table.head())
    print()
        
    print("Bottom Records of dataset")
    print(table.tail())
    print()
    
    print("Column/features/Variable  - Names of Given dataset")
    print(table.columns)
    print()
    
    print("Shape(rows x columns) - of Given dataset")
    print(table.shape)
    print()
    
    print("Data types - Given Column Names")
    print(table.dtypes)
    print()
    
    print("Summry of dataset")
    print(table.info())
    print()
    
    print("To see the count of null/nan values in columns of dataset")
    print(table.isnull().value_counts())
    print()
    
    print("Dataset Summary ")
    print(table.describe())
    print()
    
```

```python
basic_inspection(df)
```

# Q 1
Q. 1) Convert the Datatype of 'Date' column to Date-Time format.

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df.dtypes
```

```python
df.head()
```

# Q 2
Q. 2) Add a new column ''year'' in the dataframe, which contains years only.

(2.B) Add a new column ''month'' as 2nd column in the dataframe, which contains month only.


```python
df['year'] = df['date'].dt.year
df.insert(2,"month", df["date"].dt.month)
```

```python
df.head()
```

# Q 4
Q. 4) Show all the records where 'No. of Crimes' is 0. And, how many such records are there ?

```python
df[df["no_of_crimes"]==0]
```

# Q 5
Q. 5) What is the maximum & minimum 'average_price' per year in england ?

```python
df_england=df[df["area"]=="england"]
df_england.groupby("year").average_price.agg([min,max])
```

# Q 6
Q. 6) What is the Maximum & Minimum No. of Crimes recorded per area ?

```python
df.groupby("area")["no_of_crimes"].agg([min,max])
```

# Q 7
Q. 7) Show the total count of records of each area, where average price is less than 100000.

```python
df[df["average_price"]<100000].area.value_counts()
```

# Q 3
Q. 3) Remove the columns 'year' and 'month' from the dataframe.

```python
df.drop(columns=["year","month"],axis=1,inplace=True)
```
