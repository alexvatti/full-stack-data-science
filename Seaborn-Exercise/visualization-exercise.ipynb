## 1. Bar Plot

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
print(tips.head())
print(tips.columns)
# visualization
plt.figure(figsize=(7,5))
ax = sns.barplot(x="day", y="total_bill", data=tips)
plt.xlabel("day")
plt.ylabel("bill")
plt.title("day vs bills")
plt.show()
```

**Observations**
1. Highest to lowest:bill paid to service: Sunday,Saturday, Thursday, Friday

```python
# importing the required library
import seaborn as sns
import matplotlib.pyplot as plt
# read a titanic.csv file
# from seaborn library
titanic = sns.load_dataset('titanic')
# who v/s fare barplot
print(titanic.head())
print(titanic.columns)
sns.barplot(x = 'pclass',y = 'survived', data = titanic)
# Show the plot
plt.xlabel("Class")
plt.ylabel("Survived")
plt.title("Class vs Survived")
plt.show()
```

**Observations**
1. Survived Highest to lowest: Class 1, Class 2, Class 3

 ## 2. Pie Plot
 A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data. 

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
print(iris.head())
print(iris.columns)

flowers = iris.species.unique()
print(flowers)
print(iris.species.value_counts())

plt.pie(labels=flowers,x=iris.species.value_counts())
plt.show()
```

**Observations**
1. Iris dataset contain three equal species : 'setosa', 'versicolor', 'virginica'

```python
print(titanic.head())
print(titanic.columns)
print(titanic.survived.value_counts(normalize=True))
plt.pie(labels=["not-survived","survived"],x=titanic.survived.value_counts())
plt.show()
```

**Observations**
1. Survived =38% , Non-Survived=62%

## 3. Heatmap
Plot rectangular data as a color-encoded matrix.
1. The correlation coefficient has values between -1 to 1
2. A value closer to 0 implies weaker correlation (exact 0 implying no correlation)
3. A value closer to 1 implies stronger positive correlation
4. A value closer to -1 implies stronger negative correlation

```python
plt.figure(figsize=(7,4))
sns.heatmap(iris.corr(),annot=True)
plt.show()
```

```python
#Correlation with output variable
cor=iris.corr()
cor_target = abs(cor["petal_width"])#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.5]
relevant_features
```

**Observations**
1.  dark color shows the high correlation between the values and the light colors shows less correlation between the variables

```python
plt.figure(figsize=(10,8))
sns.heatmap(titanic.corr(),annot=True)
plt.show()
```

## 4. Box Plot
A Box and Whisker Plot (or Box Plot) is a convenient way of visually displaying the data distribution through their quartiles. 

```python
sns.set_theme(style="whitegrid")
print(tips.head())
print(tips.columns)
# visualization
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips)
```

**Observations**
1. There are outliers towards upper extreme

```python
print(iris.head())
print(iris.columns)
l = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig,ax = plt.subplots(2,2,figsize=(12,12))
fig.suptitle('Box-Plots')
for i in range(0,2):
    for j in range(0,2):
        feature_name=l.pop()
        sns.boxplot(x="species", y=feature_name, data=iris, ax=ax[i,j],orient="v")
```

**Observations**
1. Setosa w.r.t petal width has outliers towards upper extreme
2. Setosa w.r.t petal length has outliers towards upper and lower extreme
3. setosa,  virginica w.r.t sepal width has outliers towards upper and lower extreme
4. virginica w.r.t sepal length has outliers towards lower extreme
5. versicolor w.r.t petal length has outliers towards lower extreme

## 5. Pair Plot
To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function

```python
sns.set(font_scale=1.5)
sns.pairplot(data = iris, kind = "scatter", hue="species", aspect =1, height = 3)
plt.show()
```

**Observations**
1. petal length, petal width, sepal length and sepal width  - A bell curve is a common type of distribution for a variable, also known as the normal distribution.
2. petal length, petal width - distributions are separated
3. sepal length and sepal width - distributions are highly overlapped

```python
print(tips.head())
print(tips.columns)
# visualization
sns.set(font_scale=1.5)
sns.pairplot(data = tips, kind = "scatter", hue="smoker", aspect =1, height = 3)
plt.show()
```

## 6. Violin Plot
A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
fig.suptitle('Violin-Plots')
l = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i in range(0,2):
    for j in range(0,2):
        feature_name=l.pop()
        sns.violinplot(ax=axes[i, j], x='species',y=feature_name, data=iris)
plt.show()

```

## 7. countplot
method is used to Show the counts of observations in each categorical bin using bars.

```python
sns.set_theme(style="darkgrid")
sns.countplot(x="class", hue='alive', data=titanic)
plt.show()
```

**Observations**
1. First-class  Passengers: survived is more than not-survivied (survived:non-survived=2:1)
2. Second-class Passengers:  both survived and not-survived are almost same (survived:non-survived=1:1)
3. Third -class Passengers: Not survived is far more than survived(survived:non-survived=1:3)

```python
sns.countplot(x="sex", hue='alive', data=titanic)
plt.show()
```

**Observations**
1. Survived Male Passengers is less compared to female Passengers
2. Male Passengers (Survived : Non Survived=1:4)
3. Female Passengers (Survived : Non Survived=3:1)

## 8. Kde Plot
A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle('Kde-Plots')
l = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for i in range(0,2):
    for j in range(0,2):
        name=l.pop()
        sns.histplot(ax=axes[i, j], hue='species',x=name, data=iris,kde=True)
plt.show()
```

**Observations**
1. Bell Shaped Curve -  used to depict a normal distribution consists of a symmetrical bell-shaped curve.
