```python
import numpy as np
```

```python
def calculate_distance(v1,v2):
    """distance between points"""
    dist_list = []
    for i in v1:
        d = np.square(i[0]-v2[0]) + np.square(i[1]-v2[1])
        dist_list.append(np.sqrt(d))
    return(max(dist_list),min(dist_list))
```

```python
v1 = [(1,2), (3,5),(5,7)]
v2 = (10,20)
```

```python
max_dist, min_dist=calculate_distance(v1,v2)
print("max distance",max_dist)
print("min distance:",min_dist)
```
