# Matrix operationa using numpy

```python
import numpy as np
def matrix_add(a,b):
    """Matrix addition is the operation of adding two matrices of the same size to produce a new matrix. 
    The resulting matrix has the same dimensions as the original matrices."""
    if (a.shape == b.shape):
        c = a+b
        return c
    else:
        print(f"Invalid Dimentions of matrices {a.shape} and {b.shape}\n")
        return None

def matrix_sub(a,b):
    """Matrix subtraction is the process of subtracting one matrix from another. 
    This operation can be performed element-wise"""
    if (a.shape == b.shape):
        c=a-b
        return c
    else:
        print(f"Invalid Dimentions of matrices{a.shape} and {b.shape}\n")
        return None


def matrix_mul(a,b):
    """Matrix multiplication refers to the operation of combining 
    two or more matrices into a single result, called a product,"""
    if a.shape[-1] == b.shape[0] :
        c=np.dot(a,b)
        return c
    else:
        print(f"Invalid Dimentions of matrices{a.shape} and {b.shape}\n")
        return None

def matrix_inv(a):
    """The existence of an inverse matrix is guaranteed only 
    if A is non-singular, i.e., det(A) â‰  0. """
    if np.linalg.det(a) != 0:
        r=np.linalg.inv(a)
        return r
    else:
        print("Det of matrix is zero\n")
        return None


def matrix_transpose(a):
    """Matrix transpose is an operation that are used in linear algebra and mathematics
    in which the rows and columns of a matrix are switched."""
    return a.transpose()

def matrix_det(a):
    """A matrix determinant is a scalar value that measures the degree of
    linear independence of the rows or columns of a square matrix. 
    It is calculated using a specific formula based on the entries of the matrix. 
    The determinant has important properties in linear algebra, such as being zero 
    if and only if the matrix is singular"""
    
    if a.shape[0] == a.shape[1]:
        return np.linalg.det(a)
    else:
        print(f"Invalid Matrix shape {a.shape}\n")
        return None

def matrix_solve_equation(a,b):
    """solve linear equations"""
    if True:
        return np.linalg.solve(a, b)
    else:
        print("Invalid Matrix dimensions\n")
        return None

def eigen_vectors_values(a):
    """find eigen vectors and eigen values"""
    if a.shape[0] == a.shape[1]:
        eigenvalues, eigenvectors = np.linalg.eig(a)
        return eigenvalues,eigenvectors
    else:
        print("Invalid Matrix dimensions\n")
        return None
```
