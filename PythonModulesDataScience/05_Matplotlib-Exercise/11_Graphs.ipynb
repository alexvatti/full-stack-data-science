```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Graph  Y=X

```python
y = [x for x in range(1,10)]
x=y
print(f"x values", x)
print(f"y values", y)
```

```python
plt.grid()
plt.plot(x,y)
plt.show()
```

# 2. Graph  Y=mX+c

```python
x = [i for i in range(0,10)]
m=3
c=2
y = [m*i+c for i in x]
print(f"x values", x)
print(f"y values", y)
```

```python
plt.grid()
plt.plot(x,y)
plt.xticks([i for i in range(-3,10,1)])
plt.yticks([i for i in range(-5,30,3)])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# 3. Graph  Y=|X|

```python
x = [i for i in range(-10,11)]
y = [abs(i) for i in x]
print(f"x values", x)
print(f"y values", y)
```

```python
plt.grid()
plt.plot(x,y)
plt.xticks([i for i in range(-10,11,1)])
plt.yticks([i for i in range(-3,10,1)])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# 4. Graph  Y=X*X 

```python
x = [i for i in range(-10,11)]
y = [pow(i,2) for i in x]
print(f"x values", x)
print(f"y values", y)
```

```python
plt.grid()
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# 5. Graph  Y=sin(radiance)

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s =  np.sin(2 * np.pi * t)

plt.grid()
plt.plot(t,s)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# 6. Graph  Y=cos(radiance)

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s =  np.cos(2 * np.pi * t)

plt.grid()
plt.plot(t,s)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
